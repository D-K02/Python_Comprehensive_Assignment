```python
# Checking for NaN values
df[df['Name'].isna()]
df[df['Team'].isna()]
df[df['Number'].isna()]
df[df['Position'].isna()]
df[df['Age'].isna()]
df[df['Height'].isna()]
df[df['Weight'].isna()]
df[df['College'].isna()] # College column has NaN data
```

```python
# filling Nan data in Salary column with 0
df['Salary']=df['Salary'].fillna(0)
```

# Analysis Tasks:

## 1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees. (2 marks)

```python
Employees_in_each_Team =df.value_counts('Team')
Percentage_Split_Relative = Employees_in_each_Team/Employees_in_each_Team.sum()*100
Percentage_Split_Relative
```

```python

counts = df['Team'].value_counts()
counts.plot(kind='bar', color='skyblue')
plt.xlabel('Team')
plt.ylabel('No of Players')
plt.title('Distribution of employyes across various teams')
plt.show()
```

## New Orleans Pelicans  Team has the most Players.
## Minnesota Timberwolves & Orlando Magic      Teams have the least number of players

```python
plt.figure(figsize=(8,10),dpi=100)
df.value_counts('Team').plot(kind="pie",autopct='%1.1f%%', radius=2, shadow=True,)
plt.title('The percentage split relative to the total number of employees')
plt.axis('equal')
plt.show()
```

## 4. Discover which team and position have the highest salary expenditure. (2 marks)
    

```python
salary_exp_by_team = df.groupby(['Team'])[['Salary']].sum()
```

```python
Team_highest_Exp =salary_exp_by_team['Salary'].sort_values(ascending = False)
Team_highest_Exp .head()
```

```python
Team_highest_Exp.plot(kind='bar', color='black')
plt.xlabel('Salary')
plt.ylabel('Team')
plt.title('Team with salary expenditure')
plt.show()
```

# Team "Cleveland Cavaliers" have the highest salary expenditure of  106988689.0

```python
salary_exp_by_position = df.groupby(['Position','Team'])[['Salary']].max()
```

```python
Position_highest_Exp =salary_exp_by_position['Salary'].sort_values(ascending = False)
Position_highest_Exp .head()
```

```python
Position_highest_Exp.head().plot(kind='bar', color='red')
```

# The position with highest salary expenditure is "SF" for 25000000.0, which is spent by the team "Los Angeles Lakers".

## 5. Investigate if there's any correlation between age and salary, and represent it visually. (2 marks)

```python
plt.figure(figsize=(8,5),dpi=100)
sns.scatterplot(y="Salary",x="Age",hue="Position",data=df)
# plt.plot(df['Salary'], df['Age'],label='ABC Company',linestyle="--",color="black",marker="*",markeredgecolor="r",linewidth=3,markersize=8)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("ABC Company")
#show
plt.show()
```

From the above scatter plot we can infer the following
1. The predominant age  group identfied in the earlier questions is age between 23 to 27. The palyers in that age group recieves more salary than the other age groups irrespective of their position. But there are some exceptions to this case.
2. No of players with age above 30 years is too low compared to other age groups. And the salary provided to players above 30 years is in the lowest range.
3. As we can see in the scatter plot, large number of players in the pre dominant age group is also paid in lowest salary range
4. The no of players below 23 years old and above 27 years old is low when compared to predominant group.
